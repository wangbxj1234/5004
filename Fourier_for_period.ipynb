```python
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft

data = pd.read_csv('/content/HK_Meteorological_Data_Daily_1884-2021.csv')
mean_temperature = data['Mean (deg. C)'][91:]


#%% Q1
select_temperature = pd.to_numeric(mean_temperature[40000:42000])
time = range(len(select_temperature))
plt.figure(figsize=(20,10))
plt.plot(time,select_temperature)
#plt.xticks(np.arange(0,2000,step=50))
plt.show()
#   period=about 365
```

```python
pi=np.pi
x=np.arange(0,100,1)
y=np.sin((2*pi*x)/25)
ck=fft(y)
plt.xticks(np.arange(0,100,step=10))
plt.plot(x,abs(ck)**2)
```

```python
#%% Q2
N = len(mean_temperature)
fft_x = fft(mean_temperature)
ck_2 = (np.abs(fft_x))**2
plt.figure(figsize=(30,10))
x = range(500)
y = ck_2[0:len(x)]
log_y = np.log(y)
plt.xticks(np.arange(0,200,step=10))
plt.plot(x,log_y)
#Usually a peak at 0Hz in the Frequency domain indicates the presence of a DC (Constant) bias in time series data.
```

```python
N/130 #about a year
```

```python
# Q3
fft_x = fft(mean_temperature)
fft_x[int(0.1*N):]=0
y2 = ifft(fft_x)
y2=y2[40000:42000]
plt.figure(figsize=(20,10))
plt.plot(time,select_temperature,color='b')
plt.plot(time,y2,color='r')
plt.show()
```

```python
# Q4
fft_x = fft(mean_temperature)
fft_x[int(0.02*N):]=0
y2 = ifft(fft_x)
y2=y2[40000:42000]
plt.figure(figsize=(20,10))
plt.plot(time,select_temperature,color='b')
plt.plot(time,y2,color='r')
plt.show()
```
